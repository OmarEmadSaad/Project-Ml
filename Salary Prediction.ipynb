```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# import the basic libraray
```

```python
import pandas as pd
df = pd.read_csv("D:\\level 4 cs\\Smester 2\\Project ML\\Project Ml\\salary_prediction_dataset.csv")
df.head()
```

```python
# Download and read data
```

```python
print(df.info())
print(df.describe())
sns.scatterplot(data=df, x='YearsExperience', y='Salary')
```

```python
# data analysis
```

```python
X = df[['YearsExperience']]
y = df['Salary']
```

```python
# Data preparation 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Data Splitting (Training/Testing)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Data normalization
```

```python
model = KNeighborsRegressor(n_neighbors=3)
model.fit(X_train_scaled, y_train)
```

```python
# Building and training a KNN model
```

```python
y_pred = model.predict(X_test_scaled)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')
```

```python
# test 
```

```python
# Example: An employee with 5 years of experience
new_data = pd.DataFrame([[5]], columns=['YearsExperience']) 
new_data_scaled = scaler.transform(new_data)
predicted_salary = model.predict(new_data_scaled)
print(f'Predicted Salary for 5 years of experience: {predicted_salary[0]:.2f}')
```
